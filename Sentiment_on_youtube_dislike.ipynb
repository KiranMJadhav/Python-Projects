```python
import matplotlib.pyplot as plt



fig,ax=plt.subplots(figsize=(8,4))
ax.scatter(df['view_count'],df['dislikes'])
ax.set_xlabel('view of videos')
plt.ylabel('dislikes of videos')
plt.title('Relation between views and dislikes')

plt.show()
```

* We can see the scatter plot of views of videos and dislikes of videos.
* We can see the difference between this two column that there is 
  not a difference that higher the views of the video would get more dislikes its not true.
* The views of the videos does not affect the dislikes of videos.
* We can see the traffic of the dots on the x axis and the dislikes are less than 0.7 on the y axis.
